Importing SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
smt=SMOTE()
```

```python
#x,y are the inputs inside smote to create artificial rows
#trainx & trainy are givng the new rows
trainx,trainy=smt.fit_resample(x,y)
```

Model-Linear Regression czo ylable is binary

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr=LogisticRegression()
for i in range(0,1000):
    x_train,y_train,x_test,y_test=train_test_split(trainx,trainy,random_state=i,test_size=.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.preditc(x_test)
    #if both training & testing acc are same it will print
    if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
        print("At random state",i,"The model performs well")
        print("At random_state:-",i)
        print("Training accuracy score is:-", round(accuracy_score(y_train,pred_train)*100,1))
        print("Testing accuracy score is:-", round(accuracy_score(y_test,pred_test)*100,1),'\n\n')
```
